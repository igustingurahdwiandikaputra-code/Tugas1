```python
list = []
```

```python
for i in range(10):
  nilai = int(input("masukan nilai: "))
  list.append(nilai)
```

```python
print("list: ", list)
print("Nilai Tertinggi:", max(list))
print("Nilai Terkecil:", min(list))
print("Nilai Rata-rata:", sum(list)/len(list))
```

```python
import matplotlib.pyplot as plt
plt.bar(range(len(list)), list)
plt.title("Grafik Nilai")
plt.xlabel("Indeks")
plt.ylabel("Nilai")
plt.show()
```
