```python
import matplotlib.pyplot as plt
import numpy as np

#input nillai
nilai_mahasiswa = []
print("Masukkan 10 nilai mahasiswa:")
for i in range(1, 11):
    while True:
        try:
            val = float(input(f"Nilai mahasiswa ke-{i}: "))
            nilai_mahasiswa.append(val)
            break
        except ValueError:
            print("Input harus berupa angka!")

#max,min,sum
nilai_tertinggi = max(nilai_mahasiswa)
nilai_terendah = min(nilai_mahasiswa)
rata_rata = sum(nilai_mahasiswa) / len(nilai_mahasiswa)

#htiung kwlulusan
lulus = [n for n in nilai_mahasiswa if n >= 60]
tidak_lulus = [n for n in nilai_mahasiswa if n < 60]
jumlah_lulus = len(lulus)
jumlah_tidak_lulus = len(tidak_lulus)

#Hasil
print("\n" + "="*30)
print(f"Nilai Tertinggi: {nilai_tertinggi}")
print(f"Nilai Terendah : {nilai_terendah}")
print(f"Rata-rata      : {rata_rata:.2f}")
print(f"Jumlah Lulus   : {jumlah_lulus}")
print(f"Tidak Lulus    : {jumlah_tidak_lulus}")
print("="*30)

#Grafik
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

#Grafik Nilai Tertinggi & Terendah
ax[0].bar(['Terendah', 'Tertinggi'], [nilai_terendah, nilai_tertinggi], color=['red', 'green'])
ax[0].set_title('Perbandingan Nilai Ekstrem')
ax[0].set_ylabel('Skor')

#Grafik Kelulusan
labels = ['Lulus (>=60)', 'Tidak Lulus (<60)']
sizes = [jumlah_lulus, jumlah_tidak_lulus]
ax[1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'salmon'])
ax[1].set_title('Persentase Kelulusan Mahasiswa')

plt.tight_layout()
plt.show()
```
